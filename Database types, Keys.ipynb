#### Properties of an ideal Database
- Integrity
- Availability
- Security
- Independent from Application
- Concurrency

**Types of Databases**:

(a) Relational Databases : Tabels with rows and columns -> MySQL, Oracle, PostgreSQL etc.

(b) NoSQL Databases : For unstructured or semi-structured data such as images, documents, or videos -> MongoDB

(c) Column Databases : Store data in columns rather than rows, well suited for data warehousing and analytical applications -> Amazon Redshift, Google BigQuery

(d) Graph Databases : Used to store and query graph-structured Databases -> Neo4j, Amazon Neptune

(e) Key-Value Databases : Store data as a collection of keys and values, well suited for caching and simple data storage needs -> Redis and Amazon DynamoDB

#### Database Keys
A key in a database is an `attribute` or `set of attributes` that `uniquely identifies a tuple`(row) in a table.



1. **Super Key**:
   `Combination of columns` that uniquely identifies any row within a RDBMS table. In the above table: `roll no`, `email`, `roll + name`, `roll + branch`, `roll + email`, `name + branch + email`, `name + branch + roll`, `roll + branch + name + email`, all these set can be used to uniquely identify any row within the table.
   
3. **Candidate Key**:
   `Minimal super key`, has no redundant attributes. `Smallest set of attributes` that can be used to uniquely identify a tuple(row) in the table. In other words it has no redundant attributes. `roll no`, `email`, `roll + name`(because roll no can independently do the thing then why to make roll+name), `roll + branch `, `name + branch + email`

3. **Primary Key**: `unique identifier` for each tuple in a table. A table can have only be `one primary key` and it `cannot contain null` values. `roll no`,`email`(people sometimes change it with time,it is also long not as small as roll no)



4. **Alternate Key**:
   Candidate key - Primary key = Alternate key

5. **Composite key**:
   `Primary key` made up of `two or more attributes`. Used `when a single attribute` is `not sufficient to uniquely identify a tuple` in a table.
   


6. **Surogate Key**:
   A surrogate key is a key which does not have any contextual or business meaning. Like: S.no. or emp. id. etc. Often `used when the natural key is not unique or when it is subject to change`.

   
8. **Foreign Key**:
   A `primary key` from one table that is `used to establish a relationship with another table`.
